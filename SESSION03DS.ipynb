# 1) Problems Based On Loops In Python :

PROBLEM NUMBER 1 : -

Sequence Sum :

1/1! + 2/2! + 3/3! +......

```python
# Printing the sum for the type of the sequence given above upto 'n' terms :

import math

# Taking input from the user
num = int(input("Enter the number of terms: "))

print("\nSequence (n/n!) is:\n")

# Declare a sum variable to calculate the total sum .
sum = 0

# Loop to calculate and display the sequence
for n in range(1, num + 1):
    term = n / math.factorial(n)
    sum += term
    print(f"{n}/{n}! = {term}")

print(sum)
```

Nested Loops : -

Nested loops in Python are loops inside other loops, allowing execution of repetitive tasks in multiple dimensions (like rows and columns). The inner loop runs completely for each iteration of the outer loop. Nested loops are often used in pattern printing, working with matrices, and solving complex problems that require multi-level iteration. They can be combinations of for loops, while loops, or a mix of both, depending on the problem.



---



Examples for Nested Loops :

```python
# Example - 1 :
# Multiplication Table :

for i in range(1, 4):        # Outer loop
    for j in range(1, 4):    # Inner loop
        print(i * j, end="\t")
    print()

# Example - 2 :
# Pattern - Printing :

rows = 5
for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print("*", end=" ")
    print()


# Example - 3 :
# Matrix - Addition :

# Adding two 2x2 matrices
A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]
result = [[0, 0], [0, 0]]

for i in range(2):
    for j in range(2):
        result[i][j] = A[i][j] + B[i][j]

# Display the result
for row in result:
    print(row)
```

Python Patterns Based on Numbers :

```python
# Pattern - 1 :
# Pyramid Pattern :

def pyramid_pattern(n):
    for i in range(1, n+1):
        # Print leading spaces
        print(' ' * (n - i), end='')
        # Print numbers
        for j in range(1, i+1):
            print(j, end=' ')
        print()

# Example usage
pyramid_pattern(5)
print()

# Pattern - 2 :
# Inverted Pyramid With Same Numbers :

def inverted_pyramid(n):
    for i in range(n, 0, -1):
        # Print leading spaces
        print(' ' * (n - i), end='')
        # Print numbers
        for j in range(1, i+1):
            print(i, end=' ')
        print()

# Example usage
inverted_pyramid(5)
print()

# Pattern - 3 :
# Diamond Pattern Using Numbers :

def diamond_pattern(n):
    # Upper half
    for i in range(1, n+1):
        print(' ' * (n - i), end='')
        for j in range(1, i+1):
            print(j, end=' ')
        print()

    # Lower half
    for i in range(n-1, 0, -1):
        print(' ' * (n - i), end='')
        for j in range(1, i+1):
            print(j, end=' ')
        print()

# Example usage
diamond_pattern(5)
print()

# Pattern - 4 :
# Floyd's Triangle Pattern Code in Python :

def floyds_triangle(n):
    num = 1
    for i in range(1, n+1):
        for j in range(1, i+1):
            print(num, end=' ')
            num += 1
        print()

# Example usage
floyds_triangle(5)
print()

# Pattern - 5 :
# Pascal's Triange Code Using Python :

def pascals_triangle(n):
    for line in range(1, n+1):
        C = 1
        for i in range(1, line+1):
            print(C, end=' ')
            C = C * (line - i) // i
        print()

# Example usage
pascals_triangle(5)
print()
```

Pattern Printing in Python :

Pattern printing in Python uses nested loops to display sequences of characters or numbers in structured forms like triangles, pyramids, or diamonds. The outer loop controls rows, while inner loops manage columns or spaces. Logical conditions help in shaping patterns, enabling creative designs using stars, numbers, or alphabets efficiently.

QUESTION : Print the following pattern using python code -

1

121

12321

1234321

```python
# Solution to the question :

rows = 4  # Number of rows

for i in range(1, rows + 1):
    # Increasing sequence
    for j in range(1, i + 1):
        print(j, end="")

    # Decreasing sequence
    for j in range(i - 1, 0, -1):
        print(j, end="")

    # Move to next line
    print()
```

# 2) Strings In Python :

Strings are one of the most fundamental data types in Python, used to represent text. They are sequences of Unicode characters enclosed in single (' '), double (" "), or triple (''' ''' or """ """) quotes.



---



1) Creating Strings :

Strings can be created in multiple ways:


```python
# Single quotes
str1 = 'Hello, World!'

# Double quotes
str2 = "Python Programming"

# Triple quotes (for multi-line strings)
str3 = '''This is a
multi-line
string.'''

# Escape characters
str4 = "He said, \"Python is awesome!\""

print(str1)
print(str2)
print(str3)
print(str4)
```



---



2) String Indexing and Slicing :

Strings are immutable (cannot be changed) and support indexing and slicing.

Example: Accessing Characters

```python
# Example Involving String Indexes and Slicing :

text = "Python"

# Indexing (starts at 0)
print(text[0])  # 'P'
print(text[-1]) # 'n' (negative index starts from the end)

# Slicing (start:stop:step)
print(text[1:4])  # 'yth' (from index 1 to 3)
print(text[:3])   # 'Pyt' (from start to index 2)
print(text[2:])   # 'thon' (from index 2 to end)
print(text[::2])  # 'Pto' (every 2nd character)
print(text[::-1]) # 'nohtyP' (reverse string)
```

3) String Methods (Common Operations) :

Python provides many built-in string methods for manipulation.


```python
# Example :

s = "hello world"
print(s.upper())       # "HELLO WORLD"
print(s.title())       # "Hello World"
print(s.capitalize())  # "Hello world"
```

7) Commonly Used String Methods :

| **Method**      | **Description**                                 | **Example**                           |
| --------------- | ----------------------------------------------- | ------------------------------------- |
| `len()`         | Returns length of string                        | `len("Hello") → 5`                    |
| `upper()`       | Converts to uppercase                           | `"hello".upper() → HELLO`             |
| `lower()`       | Converts to lowercase                           | `"HELLO".lower() → hello`             |
| `title()`       | Converts first letter of each word to uppercase | `"hello world".title() → Hello World` |
| `capitalize()`  | Converts first letter of string to uppercase    | `"python".capitalize() → Python`      |
| `strip()`       | Removes spaces from both ends                   | `" hello ".strip() → hello`           |
| `lstrip()`      | Removes spaces from left                        | `" hello".lstrip() → hello`           |
| `rstrip()`      | Removes spaces from right                       | `"hello ".rstrip() → hello`           |
| `replace(a,b)`  | Replaces substring a with b                     | `"hello".replace("l","x") → hexxo`    |
| `split()`       | Splits string into list                         | `"a,b,c".split(",") → ['a','b','c']`  |
| `join()`        | Joins list elements with a separator            | `"-".join(['a','b']) → a-b`           |
| `find()`        | Returns index of first occurrence               | `"hello".find("l") → 2`               |
| `index()`       | Like `find()` but gives error if not found      | `"hello".index("e") → 1`              |
| `count()`       | Counts occurrences of a substring               | `"banana".count("a") → 3`             |
| `startswith()`  | Checks if string starts with given substring    | `"hello".startswith("he") → True`     |
| `endswith()`    | Checks if string ends with given substring      | `"hello".endswith("o") → True`        |
| `isalnum()`     | True if string has only letters and numbers     | `"abc123".isalnum() → True`           |
| `isalpha()`     | True if string has only letters                 | `"abc".isalpha() → True`              |
| `isdigit()`     | True if string has only digits                  | `"123".isdigit() → True`              |
| `swapcase()`    | Swaps uppercase to lowercase and vice versa     | `"PyThOn".swapcase() → pYtHoN`        |
| `center(width)` | Centers string in given width                   | `"hi".center(6) → '  hi  '`           |
