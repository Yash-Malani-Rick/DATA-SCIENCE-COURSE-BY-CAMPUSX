c) Constructors In Python : -

In Python, a constructor is a special method used to initialize an object‚Äôs attributes at the time of creation. It is defined using the __init__() method and is called automatically whenever an object is instantiated. The self parameter represents the current instance of the class. Constructors can be default constructors (no parameters, assigning fixed values) or parameterized constructors (accepting arguments to set attributes dynamically). They improve code readability, reduce redundancy, and ensure that every object starts in a valid state. Python does not have multiple constructors but can mimic them using default arguments or class methods.

````python
# Coding Examples Involving Classes and Objects : -

# Banking Application / ATM Machine :
class BankAccount:
    def __init__(self, account_number, account_holder, balance=0):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = balance

    # Method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"‚úÖ Deposited: ‚Çπ{amount}. New Balance: ‚Çπ{self.balance}")
        else:
            print("‚ùå Deposit amount must be positive!")

    # Method to withdraw money
    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"‚úÖ Withdrawn: ‚Çπ{amount}. Remaining Balance: ‚Çπ{self.balance}")
        else:
            print("‚ùå Insufficient funds or invalid amount!")

    # Method to check balance
    def check_balance(self):
        print(f"üí∞ Available Balance: ‚Çπ{self.balance}")

# -------------------------------
# ATM Simulation
# -------------------------------
def atm_menu(account):
    while True:
        print("\n=== ATM MENU ===")
        print("1. Deposit")
        print("2. Withdraw")
        print("3. Check Balance")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            amount = float(input("Enter amount to deposit: "))
            account.deposit(amount)
        elif choice == '2':
            amount = float(input("Enter amount to withdraw: "))
            account.withdraw(amount)
        elif choice == '3':
            account.check_balance()
        elif choice == '4':
            print("‚úÖ Thank you for using our ATM!")
            break
        else:
            print("‚ùå Invalid option! Try again.")

# Creating an account object
user_account = BankAccount("12345678", "Yash Malani", 1000)

# Starting ATM simulation
atm_menu(user_account)


````

````python
# Code to make a fraction class : -

from math import gcd

class Fraction:
    # Constructor
    def __init__(self, numerator, denominator):
        if denominator == 0:
            raise ValueError("Denominator cannot be zero.")
        self.numerator = numerator
        self.denominator = denominator
        self.simplify()

    # Method to simplify the fraction
    def simplify(self):
        common = gcd(self.numerator, self.denominator)
        self.numerator //= common
        self.denominator //= common

    # Method to display fraction
    def display(self):
        print(f"{self.numerator}/{self.denominator}")

    # Operator Overloading for Addition
    def __add__(self, other):
        num = self.numerator * other.denominator + other.numerator * self.denominator
        den = self.denominator * other.denominator
        return Fraction(num, den)

    # Operator Overloading for Subtraction
    def __sub__(self, other):
        num = self.numerator * other.denominator - other.numerator * self.denominator
        den = self.denominator * other.denominator
        return Fraction(num, den)

    # Operator Overloading for Multiplication
    def __mul__(self, other):
        num = self.numerator * other.numerator
        den = self.denominator * other.denominator
        return Fraction(num, den)

    # Operator Overloading for Division
    def __truediv__(self, other):
        num = self.numerator * other.denominator
        den = self.denominator * other.numerator
        return Fraction(num, den)

# ----------------------------
# Testing Fraction Class
f1 = Fraction(2, 4)
f2 = Fraction(3, 5)

print("Fraction 1:", end=" "); f1.display()
print("Fraction 2:", end=" "); f2.display()

result = f1 + f2
print("Addition Result:", end=" "); result.display()

result = f1 - f2
print("Subtraction Result:", end=" "); result.display()

result = f1 * f2
print("Multiplication Result:", end=" "); result.display()

result = f1 / f2
print("Division Result:", end=" "); result.display()
````

Reference Variables : -

In Python, a reference variable is a variable that points to the memory location of an object rather than storing the object‚Äôs value directly. Multiple reference variables can refer to the same object, meaning changes made through one reference affect all others pointing to that object.

Key Points about Reference Variables : -

* Python variables store references, not the actual object data.

* Assigning one variable to another does not copy the object, only the reference.

* Useful for passing objects to functions without duplication.

* Can lead to shared modifications if multiple variables point to the same mutable object.



````python
list1 = [10, 20, 30]
list2 = list1  # list2 is a reference to list1

list2.append(40)
print(list1)  # Output: [10, 20, 30, 40] (affected by list2)
````

````python
# Checking References (id() function) :

print(id(list1))
print(id(list2))  # Both have the same memory reference
````

Pass By Reference : -

In Python, pass by reference means when an object reference (not the actual value) is passed to a function, allowing the function to modify the original object. Since Python variables store references, mutable objects (like lists, dictionaries, sets) exhibit pass-by-reference behavior.



Key Points : -

* Function receives a reference to the object, not a copy.

* Changes inside the function affect the original object (if mutable).

* Immutable objects (like integers, strings, tuples) behave differently since they cannot be altered in place.

````python
# Code Example : -

def modify_list(lst):
    lst.append(100)
    print("Inside function:", lst)

my_list = [1, 2, 3]
modify_list(my_list)
print("Outside function:", my_list)
````

‚úÖ Both lists changed since the function worked on the same reference.

Mutability of Object : -

In Python, mutability of an object refers to whether the value of the object can be changed after it is created.



1Ô∏è‚É£ Mutable Objects : -

* Their content can be modified without changing their memory reference.

* Examples: list, dictionary, set, bytearray.



````python
# Code Example : -

my_list = [1, 2, 3]
print(id(my_list))     # Memory reference
my_list.append(4)      # Modify content
print(my_list)         # [1, 2, 3, 4]
print(id(my_list))     # Same reference
````

2Ô∏è‚É£ Immutable Objects : -

* Their value cannot be changed once created; any modification creates a new object in memory.

* Examples: int, float, string, tuple, frozenset.


````python
# Code Example : -

x = "Hello"
print(id(x))
x = x + " World"       # Creates a new object
print(x)               # Hello World
print(id(x))           # Different memory reference
````

‚úÖ Key Points : -

* Mutability affects pass-by-reference behavior in functions.

* Mutable objects can lead to unexpected side effects if shared across references.

* Use immutable objects when data should not change for safety and reliability.

Encapsulation : -  

Encapsulation is one of the four pillars of Object-Oriented Programming (OOP). It is the concept of wrapping data (attributes) and methods (functions) into a single unit (class) and restricting direct access to some components of an object to protect its data integrity.

In Python, encapsulation is implemented using:

* Public Members: Accessible from anywhere.

* Protected Members: Marked with _ (single underscore); should not be accessed directly outside the class (convention).

* Private Members: Marked with __ (double underscore); cannot be accessed directly outside the class.

This ensures data hiding, allowing controlled access to data using getter and setter methods.

````python
# Code Example : -

class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder   # Public attribute
        self._account_type = "Savings"         # Protected attribute
        self.__balance = balance               # Private attribute

    # Public method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"‚úÖ Deposited ‚Çπ{amount}. New Balance: ‚Çπ{self.__balance}")
        else:
            print("‚ùå Invalid deposit amount!")

    # Public method to withdraw money
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"‚úÖ Withdrawn ‚Çπ{amount}. Remaining Balance: ‚Çπ{self.__balance}")
        else:
            print("‚ùå Insufficient balance or invalid amount!")

    # Getter method for private attribute
    def get_balance(self):
        return self.__balance

    # Setter method for private attribute
    def set_balance(self, amount):
        if amount >= 0:
            self.__balance = amount
        else:
            print("‚ùå Balance cannot be negative!")

# Creating object
acc = BankAccount("Yash Malani", 5000)

# Accessing public attribute
print("Account Holder:", acc.account_holder)

# Accessing protected attribute (not recommended directly)
print("Account Type:", acc._account_type)

# Accessing private attribute directly (will cause error)
# print(acc.__balance)  # ‚ùå AttributeError

# Correct way using getter
print("Current Balance:", acc.get_balance())

# Deposit and Withdraw
acc.deposit(1000)
acc.withdraw(3000)

# Modify balance using setter
acc.set_balance(8000)
print("Updated Balance:", acc.get_balance())
````

‚úÖ Key Points of Encapsulation : -

* Data Hiding: Prevents direct modification of sensitive attributes.

* Controlled Access: Getter and setter methods ensure safe data manipulation.

* Improved Security: Protects object‚Äôs state from accidental or unauthorized changes.

* Reusability: Encapsulation makes classes more modular and reusable.

Static Variables : -

In Python, static variables (also known as class variables) are shared by all objects of a class. They are declared inside the class but outside all methods, meaning they belong to the class itself rather than any specific object. Any change made to a static variable affects all instances of that class.

üìå Key Points : -

* Declared directly under the class definition (not inside __init__).

* Shared among all objects of the class.

* Accessed using either the class name or the object reference.

* Useful for storing data common to all objects, such as a counter or organization name.

````python
# Code Example : -

class Student:
    school_name = "DPS"   # Static/Class variable

    def __init__(self, name, roll):
        self.name = name          # Instance variable
        self.roll = roll

# Creating objects
s1 = Student("Yash", 101)
s2 = Student("Aman", 102)

# Accessing static variable
print(s1.school_name)      # DPS
print(s2.school_name)      # DPS
print(Student.school_name) # DPS

# Changing static variable using class name
Student.school_name = "Delhi Public School"

print(s1.school_name)  # Delhi Public School
print(s2.school_name)  # Delhi Public School
````

Key Differences from Instances Variables : -

| Feature     | Static Variable (Class Variable) | Instance Variable        |
| ----------- | -------------------------------- | ------------------------ |
| Scope       | Shared by all objects            | Unique to each object    |
| Declaration | Inside class but outside methods | Inside `__init__` method |
| Access      | Class name or object reference   | Object reference only    |

Class Relationship : -

In Object-Oriented Programming (OOP), Class Relationships describe how two or more classes are connected and interact with each other. These relationships help in designing structured, reusable, and scalable systems.

1Ô∏è‚É£ Types of Class Relationships :

1. Association : -

  * A basic relationship where one class uses another.

  * Shows a "uses-a" or "has-a" connection.


````python
# Code Example : -

class Driver:
    def __init__(self, name):
        self.name = name

class Car:
    def __init__(self, model, driver):
        self.model = model
        self.driver = driver

d = Driver("Yash")
c = Car("Toyota", d)   # Car is associated with Driver
````

2. Aggregation : -

    * A weaker form of association where one class contains another, but both can exist independently.

````python
# Code Example : -

class Department:
    def __init__(self, name):
        self.name = name

class Employee:
    def __init__(self, name, dept):
        self.name = name
        self.dept = dept

dept = Department("HR")
emp = Employee("Aman", dept)  # Employee has a Department, but both can exist separately
````

 3. Composition : -

    * A strong form of aggregation, where one class cannot exist without the other.

````python
# Code Example : -

class Engine:
    def __init__(self, power):
        self.power = power

class Car:
    def __init__(self):
        self.engine = Engine("1500cc")  # Engine's lifecycle depends on Car

c = Car()  # Destroying Car destroys Engine
````

4. Inheritance : -
    * A "is-a" relationship, where a subclass inherits properties and methods of a parent class.

````python
# Code Example : -

class Animal:
    def eat(self):
        print("This animal eats food")

class Dog(Animal):   # Dog is an Animal
    def bark(self):
        print("Dog barks")
````

5. Dependency : -
    * A temporary relationship, where one class depends on another to perform a task.



````python
# Code Example : -

class Printer:
    def print_doc(self, document):
        print("Printing:", document)

class User:
    def use_printer(self):
        p = Printer()
        p.print_doc("Report")  # User depends on Printer temporarily
````

Summary : -


| Relationship | Meaning           | Lifespan Dependency    |
| ------------ | ----------------- | ---------------------- |
| Association  | Uses/Has-a        | Independent            |
| Aggregation  | Has-a (container) | Independent            |
| Composition  | Has-a (strong)    | Dependent              |
| Inheritance  | Is-a              | Parent-child hierarchy |
| Dependency   | Uses temporarily  | Independent            |

Aggregation In OOP : -

Aggregation is a type of association relationship in Object-Oriented Programming where one class (the whole) contains a reference to another class (the part). Both classes can exist independently, meaning that the "part" object can live on its own, even if the "whole" object is destroyed.

It represents a "Has-A" relationship but with loose coupling compared to composition.

Key Characteristics : -

* A class is part of another class, but can exist independently.

* Both objects have separate lifecycles.

* Shown in UML diagrams with a hollow diamond arrow.

````python
# Code Example : -

class Department:
    def __init__(self, name):
        self.name = name

class Employee:
    def __init__(self, emp_name, department):
        self.emp_name = emp_name
        self.department = department   # Aggregation (Employee has a Department)

    def show_details(self):
        print(f"Employee: {self.emp_name}, Department: {self.department.name}")

# Creating department object
dept1 = Department("Human Resources")

# Creating employee object and associating with department
emp1 = Employee("Yash", dept1)

emp1.show_details()  # Output: Employee: Yash, Department: Human Resources

# Deleting employee object does not affect department
del emp1
print(dept1.name)    # Output: Human Resources
````

Key Differences : -

| Feature    | Aggregation                    | Composition                           |
| ---------- | ------------------------------ | ------------------------------------- |
| Dependency | Weak (objects independent)     | Strong (part depends on whole)        |
| Lifespan   | "Part" can outlive the "whole" | "Part" destroyed when "whole" is gone |
| UML Symbol | Hollow diamond                 | Filled diamond                        |

Inheritance : -

Inheritance is one of the fundamental concepts of Object-Oriented Programming (OOP) that allows a class (child or derived class) to acquire properties and behaviors (attributes and methods) of another class (parent or base class).

It promotes code reusability, modularity, and hierarchical relationships between classes.
The derived class can also extend or override the behavior of the base class.

Key Points In Inheritance : -

* Enables code reusability by avoiding duplicate code.

* Provides a "is-a" relationship (e.g., A Dog is an Animal).

* Allows method overriding in the child class.

* Supports different types of inheritance:

 * Single Inheritance (one parent ‚Üí one child)

 * Multiple Inheritance (multiple parents ‚Üí one child)

 * Multilevel Inheritance (parent ‚Üí child ‚Üí grandchild)

 * Hierarchical Inheritance (one parent ‚Üí multiple children)

 * Hybrid Inheritance (combination of types)



````python
# Example for Single Inheritance : -

# Base class (Parent)
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} is eating.")

# Derived class (Child)
class Dog(Animal):
    def bark(self):
        print(f"{self.name} is barking.")

# Creating object of child class
d = Dog("Tommy")
d.eat()    # Inherited from Animal
d.bark()   # Defined in Dog


# Example for Method Over-Riding : -

class Animal:
    def sound(self):
        print("Animals make sounds.")

class Cat(Animal):
    def sound(self):
        print("Cat meows.")  # Overriding parent method

obj = Cat()
obj.sound()

````

Advantages of Inheritance : -  

* Promotes code reusability and organization.

* Helps in creating a logical class hierarchy.

* Supports polymorphism (one name, multiple behaviors).

* Reduces redundancy and makes code more maintainable.

Super Keyword : -  

In Python, the super() keyword is used to access methods and attributes of a parent (base) class from its child (derived) class. It allows you to reuse and extend the functionality of the parent class without directly referencing its name, making the code more maintainable and flexible.



Key Points : -

* Used to call parent class methods or constructors.

* Avoids hardcoding the parent class name, making code adaptable in multi-level inheritance.

* Commonly used to invoke the parent‚Äôs __init__() constructor or overridden methods.



 Example: Using super() to Call Parent Constructor

````python
class Parent:
    def __init__(self, name):
        self.name = name
        print("Parent Constructor Called")

class Child(Parent):
    def __init__(self, name, age):
        super().__init__(name)   # Calls Parent class constructor
        self.age = age
        print("Child Constructor Called")

c = Child("Yash", 21)
print(c.name, c.age)
````

Example: Using super() with Method Overriding

````python
class Animal:
    def sound(self):
        print("Animals make sounds")

class Dog(Animal):
    def sound(self):
        super().sound()       # Call parent method
        print("Dogs bark")

d = Dog()
d.sound()
````

Benefits of super() : -

* Promotes code reusability.

* Supports multiple inheritance by using Method Resolution Order (MRO).

* Improves readability and maintainability of class hierarchies.

Super Constructor : -

In Python, when a child class inherits from a parent class, it often needs to initialize attributes defined in the parent class. To achieve this, the super() constructor is used to call the parent class's __init__() method.

This ensures that all the properties of the parent class are properly initialized in the child class, avoiding code duplication. It also maintains inheritance hierarchy in cases of multi-level or multiple inheritance.

Key Concepts : -

* super().__init__() calls the parent class constructor.

* Used in single and multiple inheritance scenarios.

* Ensures constructor chaining, meaning parent attributes are initialized before child-specific attributes.

* Prevents the need to hardcode parent class name, making the code more maintainable.

````python
# Code Example : -

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print("Parent Constructor: Person initialized")

class Student(Person):
    def __init__(self, name, age, roll_no):
        super().__init__(name, age)  # Calling parent constructor
        self.roll_no = roll_no
        print("Child Constructor: Student initialized")

# Creating an object of Student
s1 = Student("Yash", 21, 101)
print("Name:", s1.name)
print("Age:", s1.age)
print("Roll No:", s1.roll_no)
````

How It Works :

* The child class Student constructor first calls super().__init__(), which runs the parent (Person) constructor.

* The parent class initializes name and age.

* The child class then initializes its own attribute roll_no.

* This ensures that all inherited attributes are available to the child class without rewriting parent code.

In short : The super constructor in Python allows constructor chaining, improving code reusability, scalability, and clarity in inheritance-based programs.

Using Super Outside class : -

In Python, the super() function is primarily designed to be used inside a class hierarchy, typically within instance methods or constructors, to call methods from the parent (superclass).

However, using super() outside a class definition is not standard practice because it requires a proper context (like a method call on an instance) to determine the method resolution order (MRO).

Behaviour of super() outside a class : -

1. Normally Used Inside a Class :

````python
class A:
    def show(self):
        print("Class A method")

class B(A):
    def show(self):
        super().show()   # ‚úÖ Works correctly here
        print("Class B method")

obj = B()
obj.show()
````

2. Using super() outside a class :

  * When used outside a class, you must explicitly provide arguments:
super(CurrentClass, instance).method()


````python
class Parent:
    def display(self):
        print("Parent display")

class Child(Parent):
    def display(self):
        print("Child display")

obj = Child()

# Using super outside the class
super(Child, obj).display()  # Explicitly specifying class and object
````

‚úÖ Key Points : -

* super() without arguments only works inside class methods, where Python can determine the current class and instance automatically.

* When using super() outside a class, you must pass two arguments:

  * The class name whose parent you want to access.

  * The object (instance) through which you want to call the parent method.

Can the super keyword access the data of the parent class ??

Yes, the super() keyword can access the data (attributes and methods) of the parent class, but it can only do so through the parent class's methods or constructor, not directly.

This is because super() is used to call the parent‚Äôs implementation, allowing inherited attributes to be initialized or accessed indirectly.

1.  Accessing Parent Class Data via super() in Constructor : -

````python
class Parent:
    def __init__(self, name):
        self.name = name

class Child(Parent):
    def __init__(self, name, age):
        super().__init__(name)     # Calls parent constructor
        self.age = age

obj = Child("Yash", 21)
print(obj.name)  # ‚úÖ Data from Parent accessed via super()
````

2.  Accessing Parent Class Methods (That Use Data) : -  



````python
class Parent:
    def __init__(self, value):
        self.value = value

    def show_data(self):
        print("Parent Data:", self.value)

class Child(Parent):
    def show_data(self):
        super().show_data()  # ‚úÖ Calls parent method
        print("Child has extra data.")

obj = Child(100)
obj.show_data()
````

What super() CANNOT Do Directly : -  
* You cannot directly use super().attribute_name to access data unless that attribute is:

  * Initialized in the parent‚Äôs constructor and

  * Accessible through self or a parent method.

````python
class A:
    def __init__(self):
        self.data = "Hello"

class B(A):
    def show(self):
        # print(super().data)  # ‚ùå AttributeError
        print(self.data)       # ‚úÖ Works, since data is inherited

obj = B()
obj.show()
````

‚úÖ Conclusion:

* super() can access parent class data indirectly by:

  * Calling the parent constructor (__init__).

  * Calling parent methods that return or display the data.

* Direct attribute access via super().attribute is not supported.

SINGLE INHERITANCE :

Single inheritance is an OOP concept where a child class derives properties and methods from a single parent class. It allows code reusability by enabling the subclass to access the parent‚Äôs attributes and behaviors while also adding its own specific features. This improves program modularity and avoids redundancy.

````python
# Parent Class
class Animal:
    def eat(self):
        print("This animal eats food.")

# Child Class
class Dog(Animal):
    def bark(self):
        print("Dog barks loudly.")

# Object of Child Class
d = Dog()
d.eat()   # Inherited method
d.bark()  # Child's own method
````

---

MULTI - LEVEL INHERITANCE :

Multilevel inheritance occurs when a class inherits from another class, which in turn inherits from a third class, forming a parent ‚Üí child ‚Üí grandchild hierarchy. It allows the lowest-level class to reuse properties and methods of all its ancestor classes, improving code reusability and structure.

````python
class Grandparent:
    def feature1(self):
        print("Feature 1 from Grandparent")

class Parent(Grandparent):
    def feature2(self):
        print("Feature 2 from Parent")

class Child(Parent):
    def feature3(self):
        print("Feature 3 from Child")

# Creating object of Child class
obj = Child()
obj.feature1()  # Inherited from Grandparent
obj.feature2()  # Inherited from Parent
obj.feature3()  # Defined in Child
````

---

HIERARCHICAL INHERITANCE :

Hierarchical inheritance is a type of OOP inheritance where multiple child classes inherit from a single parent class. It allows code reusability, as the common attributes and methods are defined once in the parent class and shared by all child classes. Each child class can have its own unique properties and methods, in addition to the inherited ones. This structure is useful when different entities share common behavior but also require specialized functionalities.

````python
# Parent Class
class Vehicle:
    def start(self):
        print("Vehicle is starting...")

# Child Class 1
class Car(Vehicle):
    def drive(self):
        print("Car is being driven.")

# Child Class 2
class Bike(Vehicle):
    def ride(self):
        print("Bike is being ridden.")

# Objects of child classes
c = Car()
b = Bike()

c.start()   # Inherited method
c.drive()   # Child's method

b.start()   # Inherited method
b.ride()    # Child's method
````

---

MULTIPLE  INHERITANCE :

Multiple inheritance is an OOP concept where a child class inherits properties and methods from more than one parent class. It allows a class to combine functionalities of multiple base classes, promoting code reusability and flexibility. However, it may lead to ambiguity if multiple parent classes define methods with the same name. Python resolves such conflicts using the Method Resolution Order (MRO).

````python
class Father:
    def skill1(self):
        print("Father's Skill: Driving")

class Mother:
    def skill2(self):
        print("Mother's Skill: Cooking")

class Child(Father, Mother):
    def skill3(self):
        print("Child's Skill: Painting")

# Creating object
c = Child()
c.skill1()
c.skill2()
c.skill3()
````

HYBRID INHERITANCE :

Hybrid inheritance is a type of inheritance that combines two or more inheritance types (such as multiple, multilevel, or hierarchical inheritance) within the same program. It creates a more complex class relationship where a child class can inherit features from different parent classes or through multiple paths. This helps achieve code reusability and flexibility, but can sometimes lead to ambiguity (resolved using Method Resolution Order (MRO) in Python).



````python
# Base Class
class Person:
    def info(self):
        print("This is a person.")

# Derived Class 1 (Single Inheritance)
class Student(Person):
    def study(self):
        print("This person is a student.")

# Derived Class 2 (Multiple Inheritance)
class Employee(Person):
    def work(self):
        print("This person is an employee.")

# Hybrid Class (inherits from both Student and Employee)
class WorkingStudent(Student, Employee):
    def details(self):
        print("This person studies and works.")

# Creating object
ws = WorkingStudent()
ws.info()     # Inherited from Person
ws.study()    # From Student
ws.work()     # From Employee
ws.details()  # From WorkingStudent
````

Polymorphism : -

Polymorphism is an Object-Oriented Programming (OOP) concept that allows one name (method or operator) to have multiple forms or behaviors based on the object or data type it is used with. It enables writing flexible and reusable code, where the same function or method can work on different objects.

Python supports polymorphism in two main ways:

* Compile-time Polymorphism: Achieved using method overloading or operator overloading.

* Runtime Polymorphism: Achieved using method overriding in inheritance.



Types of Polymorphism : -

1.  Method Overriding (Runtime Polymorphism) :

  * Occurs when a child class provides a new implementation of a method inherited from the parent class.

````python
class Animal:
    def sound(self):
        print("Animals make different sounds")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

# Polymorphic behavior
for obj in (Dog(), Cat()):
    obj.sound()
````

2. Operator Overloading (Compile-time Polymorphism) :

  * Python allows operators to behave differently for different objects by defining special magic methods.

````python
class Book:
    def __init__(self, pages):
        self.pages = pages

    # Overloading '+' operator
    def __add__(self, other):
        return self.pages + other.pages

b1 = Book(100)
b2 = Book(200)
print("Total pages:", b1 + b2)
````

‚úÖ Advantages of Polymorphism :

* Increases code flexibility and reusability.

* Allows single interface, multiple implementations.

* Makes programs easier to scale and maintain.

Method Overloading : -

Method Overloading is a concept in Object-Oriented Programming where multiple methods share the same name but differ in the number or type of parameters.

Unlike some languages (like Java or C++), Python does not support traditional method overloading.
Instead, Python allows default arguments or variable-length arguments (*args, **kwargs) to achieve similar behavior, meaning a single method can handle different numbers of arguments dynamically.



Key Points :

* Only the latest defined method with the same name is recognized by Python.

* Overloading can be simulated using default or variable arguments.

* Python supports operator overloading, but not true compile-time method overloading.

````python
class MathOperations:
    def add(self, a=None, b=None, c=None):
        if a is not None and b is not None and c is not None:
            return a + b + c
        elif a is not None and b is not None:
            return a + b
        elif a is not None:
            return a
        else:
            return 0

obj = MathOperations()
print(obj.add(10))         # Output: 10
print(obj.add(10, 20))     # Output: 30
print(obj.add(10, 20, 30)) # Output: 60
````

‚úÖ Summary :

* Python does not allow multiple methods with the same name.

* Method overloading is achieved using default arguments or *args.

* Provides flexibility, allowing a single method to behave differently based on the number of arguments.

Abstraction In Python : -

Abstraction is an OOP concept that focuses on hiding implementation details and showing only the essential features of an object to the user. It allows programmers to interact with objects using a simplified interface, without worrying about their internal complexities. In Python, abstraction is typically achieved using abstract classes, interfaces (via abstract methods), and encapsulation techniques.



````python
from abc import ABC, abstractmethod

class Vehicle(ABC):  # Abstract Class
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Car engine started with key ignition.")

class Bike(Vehicle):
    def start_engine(self):
        print("Bike engine started with self-start button.")

# Using abstraction
vehicles = [Car(), Bike()]
for v in vehicles:
    v.start_engine()
````

Abstract Class In Python : -

An abstract class is a class that cannot be instantiated directly and is meant to be a blueprint for other classes. It may have abstract methods (declared but not implemented), which must be implemented by its child classes. Abstract classes help enforce a common structure across all subclasses, ensuring they implement specific behaviors.



````python
from abc import ABC, abstractmethod

class Shape(ABC):  # Abstract class
    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# obj = Shape()  # ‚ùå Cannot create object of abstract class
rect = Rectangle(10, 5)
print("Area of rectangle:", rect.area())
````
